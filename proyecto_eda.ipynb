## Análisis Exploratario de Datos (EDA)

### Cargar librerías

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Cargar los datos

```python
file_path = 'cars_dataset/Car details v3.csv'
data = pd.read_csv(file_path)
```

### Mostrar las primeras filas del dataset

```python
print(data.head())
```

### Resumen estadístico de las variables numéricas

```python
print(data.describe())
```

### Verificación de valores nulos

```python
print(data.isnull().sum())
```

### Eliminar filas con valores nulos

```python
data_cleaned = data.dropna()
```

### Visualización de la distribución del precio de venta

```python
plt.figure(figsize=(8, 6))
sns.histplot(data_cleaned['selling_price'], bins=30, color='blue')
plt.title('Distribución de Precio de Venta')
plt.xlabel('Precio de Venta (₹)')
plt.ylabel('Frecuencia')
plt.show()
```

### Visualización de la distribución de kilómetros recorridos

```python
plt.figure(figsize=(8, 6))
sns.histplot(data_cleaned['km_driven'], bins=30, color='green')
plt.title('Distribución de Kilómetros Recorridos')
plt.xlabel('Kilómetros Recorridos')
plt.ylabel('Frecuencia')
plt.show()
```

### Gráfico de cajas para el año del auto

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=data_cleaned['year'], color='red')
plt.title('Distribución de Año del Auto')
plt.xlabel('Año')
plt.show()
```

### Conteo de tipos de combustible

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=data_cleaned, x='fuel', palette='viridis')
plt.title('Distribución de Tipo de Combustible')
plt.xlabel('Tipo de Combustible')
plt.ylabel('Frecuencia')
plt.show()
```
